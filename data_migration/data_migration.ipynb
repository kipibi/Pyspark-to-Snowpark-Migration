Migrating the uber_data.csv from pyspark to snowflake 
- Note : Change the dbfs path accordingly

```
##Reading data from DBFS
uber_df = spark.read.csv("/FileStore/tables/uber_data.csv", header=True, inferSchema=True)
```

```
# display dataframe 
display(uber_df)
```

```
#Checking Schema of dataframe 
uber_df.printSchema()
```

```
#checking the count of data 
uber_df.count()
```

```
#writing uber_df dataframe into snowflake table uber (snowflake table is created on the go) 
# dbtable is the parameter which takes the name of snowflake table 
uber_df.write \
  .format("snowflake") \
  .options(**options) \
  .option("dbtable", "uber") \
  .mode("overwrite") \
  .save()
```
